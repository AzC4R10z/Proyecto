```python
#Importamos las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Cargamos la base de datos sucia
df=pd.read_csv('df_sucio.csv')
df
```

```python
#Vemos la cantidad de valores únicos en la columna
columnas=df.columns
columnas
```

```python
#Información general del DataFrame, en especial tipos de datos y valores nulos
print(df.info())
```

```python
#Revisamos los valores nulos de cada columna
for i in columnas:
    print(f'Columna {i} tiene:')
    print(f"{df[i].isnull().sum()} valores nulos")
    print("")
```

```python
#Conocemos los valores únicos de cada columna y así detectar valores atípicos
for i in columnas:
    print(f'Valores únicos en la columna {i}:')
    print(df[i].unique())
    print("")
```

Detectamos que varias columnas numéricas están como texto (object), así que:

```python
#Construimos un diccionario con las columnas numéricas que están como texto:
cols_numericas=[
    "Charging_Power_kW", "Charging_Time_hr", "Charge_Cycles",
    "Avg_Speed_kmh", "Max_Speed_kmh", "Acceleration_0_100_kmh_sec",
    "Resale_Value_USD", "Year"
]
#Transformamos valores no numéricos en NaN, para después decidir cómo tratarlos.
for c in cols_numericas:
    df[c] = pd.to_numeric(df[c], errors='coerce') #Según una pequeña investigación, los valores no numéricos se transforman en NaN con errors='coerce'
```

```python
#Usamos un ciclo for para limpiar los datos
for i in columnas:
    #Llenamos los valores nulos con 'Desconocido'
    df[i]=df[i].fillna('Desconocido')
    #Reemplazamos los valores atípicos por 'Desconocido'
    df[i]=df[i].replace('Ver4$zul','Desconocido')
```

```python
#Verificamos los cambios
for i in columnas:
    print(f'Valores únicos en la columna {i} después de la limpieza:')
    print(df[i].unique())
    print("")
```

```python
#Verificamos los valores nulos finales
df.isnull().sum()
```

```python
#Vemos si todas las columnas son iguales
df.info()
```

```python
#Renombramos los nombres de las columnas del DataFrame limpio
df=df.rename(columns={'Vehicle_ID': 'ID_del_Vehículo', 'Make': 'Marca', 'Model': 'Modelo', 'Year': 'Año', 'Region': 'Región',
                      'Vehicle_Type': 'Tipo_de_Vehículo', 'Battery_Capacity_kWh': 'Capacidad_de_Batería_kWh',
                      'Battery_Health_%': 'Salud_de_Batería_%', 'Range_km': 'Autonomía_km',
                      'Charging_Power_kW': 'Potencia_de_Carga_kW', 'Charging_Time_hr': 'Tiempo_de_Carga_hr',
                      'Charge_Cycles': 'Ciclos_de_Carga', 'Energy_Consumption_kWh_per_100km': 'Consumo_de_Energía_por_100km_recorridos',
                      'Mileage_km': 'Kilometraje_km', 'Avg_Speed_kmh': 'Velocidad_Promedio_kmh',
                      'Max_Speed_kmh': 'Velocidad_Máxima_kmh', 'Acceleration_0_100_kmh_sec': 'Aceleración_0_100_kmh_seg',
                      'Temperature_C': 'Temperatura_°C', 'Usage_Type': 'Tipo_de_Uso',
                        'CO2_Saved_tons': 'CO2_Ahorrado_tons', 'Maintenance_Cost_USD': 'Costo_de_Mantenimiento_USD_por_año',
                        'Insurance_Cost_USD': 'Costo_de_Seguro_USD_por_año', 'Electricity_Cost_USD_per_kWh': 'Costo_de_Electricidad_en_USD_por_kWh', 'Monthly_Charging_Cost_USD': 'Costo_Mensual_de_Carga_USD', 'Resale_Value_USD': 'Valor_de_Reventa_USD'})
```

```python
#Revisamos los valores de Región para traducirlos
print(df['Región'].unique())
print("")
#Revisamos los valores de la columna Vehicle_Type para traducirlos
print(df['Tipo_de_Vehículo'].unique())
print("")
#Revisamos los valores de la columna Usage_Type para traducirlos
print(df['Tipo_de_Uso'].unique())
print("")
```

```python
#Traducimos los valores de la columna Región
df['Región']=df['Región'].replace({'North America':'Norteamérica', 'Europe':'Europa', 'South America':'Sudamérica', 'Oceania':'Oceanía', 'Africa':'África'})
#Traducimos los valores de la columna Tipo_de_Vehículo
df['Tipo_de_Vehículo']=df['Tipo_de_Vehículo'].replace({'Sedan':'Sedán', 'Truck':'Camioneta'})
#Traducimos los valores de la columna Tipo_de_Uso
df['Tipo_de_Uso']=df['Tipo_de_Uso'].replace({'Fleet': 'Flota', 'Commercial':'Comercial', })
```
